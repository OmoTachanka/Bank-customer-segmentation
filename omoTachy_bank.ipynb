```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('sorted_df.csv')
df
```

```python
def ageRange(y):
    if 18<= y <30:
        return 'young_adults'
    elif 30<= y <60:
        return 'adults'
    elif 60<= y:
        return 'eld_b'
```

```python
df['AgeRange'] = df['CustomerAge'].apply(lambda x :ageRange(x))
df
```

```python
df.drop('Unnamed: 0',axis = 1, inplace = True)
df
```

```python
x, y  = np.unique(df['CustLocation'], return_counts = True)
x,y
```

```python
loc_df  = pd.DataFrame({'Loc':x, 'count':y})
loc_df
```

```python
loc_df = loc_df.sort_values('count', ascending = False)
loc_df = loc_df.reset_index(drop=True)
loc_df 
```

```python
top_10_loc = loc_df[:10]
top_10_loc, top_10_loc['count'].sum()
```

```python
new_df = df[df.CustLocation.isin(top_10_loc['Loc'])]
#new_df = new_df[df['TransactionType'] == 'small1']
new_df
```

```python
x, y = np.unique(new_df['TransactionDay'], return_counts = True)
x,y
```

```python
def weekdays(x):
    wd = ['Monday', 'Thursday', 'Tuesday','Wednesday']
    if x in wd:
        return 1
    else :
        return 0
new_df['WD'] = new_df['TransactionDay'].apply(lambda x:weekdays(x))
df
```

```python
x = new_df.select_dtypes(include = 'object').columns
for i in x:
    new_df[i] = new_df[i].astype('category')
new_df.dtypes
```

```python
og = new_df[['CustAccountBalance','TransactionAmount (INR)','AgeRange','AM', 'GenderCat', 'frequency', 'WD']]
```

```python
x = og.select_dtypes(include = 'category')
og = pd.get_dummies(og, columns=x.columns)
og
```

```python
og.columns
```

```python
from sklearn.preprocessing import normalize

normalized_df = normalize(og)
```

```python
og
```

```python
normalized_df.shape
```

```python
X = og.copy()

X[['CustAccountBalance', 'TransactionAmount (INR)', 'AM', 'GenderCat',\
       'frequency', 'WD', 'AgeRange_adults', 'AgeRange_eld_b',\
       'AgeRange_young_adults']] = normalized_df
normalized_df = X
```

```python
normalized_df = normalized_df.reset_index(drop = True)
```

```python
normalized_df
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit(normalized_df)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (10, 10))
plt.plot(range(1, 11), wcss, marker = 'o')
plt.show()
```

```python
kmeans_pca = KMeans(n_clusters = 3, init = 'k-means++', random_state = 42)
```

```python
kmeans_pca.fit(normalized_df)
```

```python
final_df = og.copy()
```

```python
final_df['SegmentsPCA'] = kmeans_pca.labels_
```

```python
final_df['Segments'] = final_df['SegmentsPCA'].map({0:'First',1:'Second',2:'Third'})
```

```python
final_df
```

```python
import seaborn as sns

x_axis = final_df['CustAccountBalance']
y_axis = final_df['TransactionAmount (INR)']
plt.figure(figsize = (10,8) )
sns.scatterplot(x_axis, y_axis, hue = final_df['Segments'])
plt.xlim(0,10000)
plt.ylim(0,10000)
plt.title('Clusters')
plt.show()
```
